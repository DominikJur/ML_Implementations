```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

X = np.random.rand(100, 1)

y = 2 * X + 4 + np.random.rand(100, 1)

plt.scatter(X, y, color="green")
```

```python
from linearmodel import LinReg


lr = LinReg()

lr.fit(X, y)
```

```python
x_0 = np.array((0, 1))

pred = lr.predict(x_0)

plt.scatter(X, y, color="green")
plt.plot(x_0, pred, color="red")
```

```python
lr.coef_, lr.intercept_
```
